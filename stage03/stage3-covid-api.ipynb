```python
import pandas as pd
import requests
import json
from datetime import date, datetime, timedelta
from dateutil.parser import parse
```

# Ultimos dados de cada país

```python
url_summary = "https://api.covid19api.com/summary"
response = requests.get(url_summary).text
js_summary = json.loads(response)
```

```python
df_summary = pd.DataFrame()
for country in js_summary['Countries']:
    df_summary = df_summary.append(country, ignore_index=True)
df_summary = df_summary.drop(columns=['Premium', 'Country'])
df_summary.to_csv('summary.csv', index=False)
df_summary
```

```python
url_countries = "https://api.covid19api.com/countries"
response = requests.get(url_countries).text
js_countries = json.loads(response)
```

# Primeiro caso confirmado de cada país.

```python
def get_min_date(js):
    min_date = datetime.now().date()
    for i in js:
        if parse(i['Date']).date() < min_date:
            min_date = parse(i['Date']).date()
    return min_date
```

```python
url_countries = "https://api.covid19api.com/countries"
response = requests.get(url_countries).text
js_countries = json.loads(response)

countries_dict = {'countries': [],
                  'first_case': []}
for country in js_countries:
    url = f"https://api.covid19api.com/dayone/country/{country['Slug']}/status/confirmed"
    response = requests.get(url)
    js = json.loads(response.text)
    countries_dict['countries'].append(country['ISO2'])
    countries_dict['first_case'].append(get_min_date(js))
```

```python
df_first_case = pd.DataFrame(countries_dict)
df_first_case.to_csv('first_case.csv', index=False)
df_first_case
```
