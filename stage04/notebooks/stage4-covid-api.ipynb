# Primeiro caso confirmado de cada país.

```python
countries_slug = list(map(lambda x: x["Slug"], js_countries))
```

```python
res = []
for slug in countries_slug:
    try:
        response = requests.get(f"https://api.covid19api.com/dayone/country/{slug}/status/confirmed").json()
        if len(response) >= 1 and 'Date' in response[0]:
            res.append({"country": response[0]["Country"], "country_code": response[0]['CountryCode'], "first_case": response[0]['Date']})
    except:
        pass
```

```python
res
```

```python
first_cases_collection.delete_many({})
first_case_insert = first_cases_collection.insert_many(res)
```
