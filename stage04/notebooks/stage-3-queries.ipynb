```sql
DROP TABLE IF EXISTS summary;
DROP TABLE IF EXISTS countries;
DROP TABLE IF EXISTS first_case;

CREATE TABLE countries (
    country varchar(70) not null,
    country_code varchar(2) not null,
    slug varchar(70)
) AS SELECT
    Country,
    ISO2,
    Slug
from csvread('../data/countries.csv');

CREATE TABLE first_case (
    country_code varchar(2) not null,
    first_case varchar(10),
    FOREIGN KEY(country_code)
        REFERENCES countries(country_code)
            ON DELETE NO ACTION
            ON UPDATE NO ACTION,
) AS SELECT
    countries,
    first_case
from csvread('../data/first_case.csv');

CREATE TABLE summary (
    country_code varchar(2) not null,
    new_confirmed decimal,
    new_deaths decimal,
    new_recovered decimal,
    slug varchar(70),
    total_confirmed decimal,
    total_deaths decimal,
    total_recovered decimal,
    FOREIGN KEY(country_code)
        REFERENCES countries(country_code)
            ON DELETE NO ACTION
            ON UPDATE NO ACTION,
    FOREIGN KEY(slug)
        REFERENCES countries(slug)
            ON DELETE NO ACTION
            ON UPDATE NO ACTION,
) AS SELECT
    CountryCode,
    NewConfirmed,
    NewDeaths,
    NewRecovered,
    Slug,
    TotalConfirmed,
    TotalDeaths,
    TotalRecovered
FROM csvread('../data/summary.csv');
```

```sql
select * from countries;
select * from summary;
select * from first_case;
```

## Dado determinado pais, gostariamos de saber quando ocorreu o seu primeiro caso

```sql
select c.country, f.first_case
    from countries as c
inner join 
    first_case as f on f.country_code = c.country_code
```

## Dado o nome de um pais, gostariamos de saber mais informações sobre a covid 

```sql
select c.country, s.*, f.first_case
    from countries as c
left join 
    summary as s on c.country_code = s.country_code
left join
    first_case as f on c.country_code = f.country_code
```
